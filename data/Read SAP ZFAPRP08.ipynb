```python
saptextfile='5555.xls'
saptextfile2=saptextfile+'.txt'
file1=open(saptextfile,'r',encoding='utf_16_le')
file2=open(saptextfile2,'w')
line=file1.readline()
while line != '':
    #print(line)
    if (line[0]!='0') and (line[0]!='1') and (line[0]!='2') and (line[0]!='3') and (line[0]!='#'):
        file2.write('#'+line)
    else:
        file2.write(line)
    line=file1.readline()
file1.close()
file2.close()
```
