```python
import pandas as pd
from datetime import datetime
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

saptextfile='5555.xls'

cols=[0,1,6,11,12,16,19,21]
colsname=['posting_date','doc_no','tax_no','doctype','payee_name','description','amount','collected']

dateparse = lambda dates: [datetime.strptime(d, '%d.%m.%Y') for d in dates]

df=pd.read_table(saptextfile,sep='\t',encoding='utf_16_le',engine='python',date_parser=dateparse,parse_dates=['posting_date'],header=None,usecols=cols,names=colsname,dtype={'doctype':'category','collected':'category'},skiprows=10,nrows=55)
df.doc_no=df.doc_no.astype('str')
df['amount']=df['amount'].str.replace(',','')
df['amount']=df['amount'].str.replace('.','')
df.amount=df.amount.astype('int') #ตอนที่แสดงผลต้องหารด้วย 100


#df.describe()
df
```

ทำการ remark ส่วนที่ไม่ใช่ข้อมูล

```python
file1=open(saptextfile,'r',encoding='utf_16_le')
saptextfile='5555.xls'
saptextfile2=saptextfile+'.txt'
file2=open(saptextfile2,'w')
line=file1.readline()
while line != '':
    #print(line)
    if (line[0]!='0') and (line[0]!='1') and (line[0]!='2') and (line[0]!='3') and (line[0]!='#'):
        file2.write('#'+line)
    else:
        file2.write(line)
    line=file1.readline()
file1.close()
file2.close()
```

```python
import pandas as pd
from datetime import datetime
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')


cols=[0,1,6,11,12,16,19,21]
colsname=['posting_date','doc_no','tax_no','doctype','payee_name','description','amount','collected']

dateparse = lambda dates: [datetime.strptime(d, '%d.%m.%Y') for d in dates]

df=pd.read_table(saptextfile2,sep='\t',engine='python',date_parser=dateparse,parse_dates=['posting_date'],header=None,usecols=cols,names=colsname,comment='#',dtype={'doctype':'category','collected':'category'})
df.doc_no=df.doc_no.astype('str')
df['amount']=df['amount'].str.replace(',','')
df['amount']=df['amount'].str.replace('.','')
df.amount=df.amount.astype('int') #ตอนที่แสดงผลต้องหารด้วย 100


#df.describe()
df
```
